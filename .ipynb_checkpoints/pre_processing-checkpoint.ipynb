# Raw Data

```python
import numpy as np
import matplotlib.pyplot as plt

# Age, Income
data = np.array([
    [21,  100],
    [24,  300],
    [29,  900],
    [35,  2700],
    [42,  8100],
])
labels = [f"P{i+1}" for i in range(len(data))]

p1 = data[0]

distances = [np.linalg.norm(p1 - point) for point in data]

# scatter plot
fig = plt.figure(figsize=(9,4))  

# left: distance vs age
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(data[:,0], distances)
for i in range(len(data)):
    ax1.text(data[i,0]+0.5, distances[i], labels[i], fontsize=12)
ax1.set_xlabel("Age", fontsize=14)
ax1.set_ylabel("Distance to P1", fontsize=14)
ax1.set_xlim(0, 8000)
ax1.tick_params(axis='both', which='major', labelsize=12)

# right: distance vs income
ax2 = plt.subplot(1, 2, 2)
ax2.scatter(data[:,1], distances)
for i in range(len(data)):
    ax2.text(data[i,1]+0.5, distances[i], labels[i], fontsize=12)
ax2.set_xlabel("Income", fontsize=14)
ax2.set_ylabel("Distance to P1", fontsize=14)
# ax2.set_xlim(0, 500)
ax2.tick_params(axis='both', which='major', labelsize=12)

plt.tight_layout()
plt.show()
```

# Log-Transform and Box-Cox Transform

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer

# generated data via exp distribution
np.random.seed(0)
data = np.random.exponential(scale=2.0, size=500).reshape(-1, 1)

# Log transform
data_log = np.log1p(data)

# Box-Cox transform
pt = PowerTransformer(method='box-cox', standardize=True)
data_boxcox = pt.fit_transform(data)


fig, axes = plt.subplots(1, 3, figsize=(14,4))

# raw data
axes[0].hist(data.ravel(), bins=30, color="skyblue", edgecolor="black", alpha=0.7)
axes[0].set_title("Original Data (Exponential)", fontsize=14)

# Log transform
axes[1].hist(data_log.ravel(), bins=30, color="orange", edgecolor="black", alpha=0.7)
axes[1].set_title("Log Transform", fontsize=14)

# Box-Cox transform
axes[2].hist(data_boxcox.ravel(), bins=30, color="green", edgecolor="black", alpha=0.7)
axes[2].set_title("Box-Cox Transform", fontsize=14)

for ax in axes:
    ax.tick_params(axis='both', labelsize=12)

plt.tight_layout()
plt.show()
```

# Log-Transform

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer

# Age, Income
data = np.array([
    [21,  100],
    [24,  300],
    [29,  900],
    [35,  2700],
    [42,  8100],
])
labels = [f"P{i+1}" for i in range(len(data))]

# Log transform
data_log = np.log1p(data)

p1 = data_log[0]
distances = [np.linalg.norm(p1 - point) for point in data_log]

# scatter plot
fig = plt.figure(figsize=(9,4)) 

# left: distance vs age
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(data_log[:,0], distances)
for i in range(len(data_log)):
    ax1.text(data_log[i,0]+0.5, distances[i], labels[i], fontsize=12)
ax1.set_xlabel("Age (log-transformed)", fontsize=14)
ax1.set_ylabel("Distance to P1", fontsize=14)
ax1.set_xlim(2, 10)
ax1.set_ylim(0, 6)
ax1.tick_params(axis='both', which='major', labelsize=12)

# right: distance vs income
ax2 = plt.subplot(1, 2, 2)
ax2.scatter(data_log[:,1], distances)
for i in range(len(data_log)):
    ax2.text(data_log[i,1]+0.5, distances[i], labels[i], fontsize=12)
ax2.set_xlabel("Income (log-transformed)", fontsize=14)
ax2.set_ylabel("Distance to P1", fontsize=14)
ax2.set_xlim(2, 10)
ax2.set_ylim(0, 6)
ax2.tick_params(axis='both', which='major', labelsize=12)

plt.tight_layout()
plt.show()

```

# Min-max Scaling

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Age, Income
data = np.array([
    [21,  100],
    [24,  300],
    [29,  900],
    [35,  2700],
    [42,  8100],
])
labels = [f"P{i+1}" for i in range(len(data))]

# Min-Max Scaling
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# distance to p1
p1 = data_scaled[0]
distances = [np.linalg.norm(p1 - point) for point in data_scaled]

# visualization
fig = plt.figure(figsize=(9,4))

# left：Distance vs Age (scaled)
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(data_scaled[:,0], distances)
for i in range(len(data_scaled)):
    ax1.text(data_scaled[i,0]+0.02, distances[i], labels[i], fontsize=12)
ax1.set_xlabel("Age (scaled)", fontsize=14)
ax1.set_ylabel("Distance to P1", fontsize=14)
ax1.set_xlim(0, 1.05)
ax1.set_ylim(0, max(distances)+0.2)
ax1.tick_params(axis='both', which='major', labelsize=12)

# right：Distance vs Income (scaled)
ax2 = plt.subplot(1, 2, 2)
ax2.scatter(data_scaled[:,1], distances)
for i in range(len(data_scaled)):
    ax2.text(data_scaled[i,1]+0.02, distances[i], labels[i], fontsize=12)
ax2.set_xlabel("Income (scaled)", fontsize=14)
ax2.set_ylabel("Distance to P1", fontsize=14)
ax2.set_xlim(0, 1.05)
ax2.set_ylim(0, max(distances)+0.2)
ax2.tick_params(axis='both', which='major', labelsize=12)

plt.tight_layout()
plt.show()
```

# Z-score Normalization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Age, Income
data = np.array([
    [21,  100],
    [24,  300],
    [29,  900],
    [35,  2700],
    [42,  8100],
])
labels = [f"P{i+1}" for i in range(len(data))]

# Z-score Normalization
scaler = StandardScaler()
data_z = scaler.fit_transform(data)

# distance to p1
p1 = data_z[0]
distances = [np.linalg.norm(p1 - point) for point in data_z]

# visualization
fig = plt.figure(figsize=(9,4))

# left: Distance vs Age (z-score normalized)
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(data_z[:,0], distances)
for i in range(len(data_z)):
    ax1.text(data_z[i,0]+0.05, distances[i], labels[i], fontsize=12)
ax1.set_xlabel("Age (Z-score)", fontsize=14)
ax1.set_ylabel("Distance to P1", fontsize=14)
ax1.set_xlim(min(data_z[:,0])-0.5, max(data_z[:,0])+0.5)
ax1.set_ylim(0, max(distances)+0.5)
ax1.tick_params(axis='both', which='major', labelsize=12)

# right: Distance vs Income (z-score normalized)
ax2 = plt.subplot(1, 2, 2)
ax2.scatter(data_z[:,1], distances)
for i in range(len(data_z)):
    ax2.text(data_z[i,1]+0.05, distances[i], labels[i], fontsize=12)
ax2.set_xlabel("Income (Z-score)", fontsize=14)
ax2.set_ylabel("Distance to P1", fontsize=14)
ax2.set_xlim(min(data_z[:,1])-0.5, max(data_z[:,1])+0.5)
ax2.set_ylim(0, max(distances)+0.5)
ax2.tick_params(axis='both', which='major', labelsize=12)

plt.tight_layout()
plt.show()
```

# Robust Normalization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler

# Age, Income
data = np.array([
    [21,  100],
    [24,  300],
    [29,  900],
    [35,  2700],
    [42,  8100],
])
labels = [f"P{i+1}" for i in range(len(data))]

# Robust Normalization (median & IQR)
scaler = RobustScaler()
data_robust = scaler.fit_transform(data)

# distance to p1
p1 = data_robust[0]
distances = [np.linalg.norm(p1 - point) for point in data_robust]

# visualization
fig = plt.figure(figsize=(9,4))

# left: Distance vs Age (Robust normalized)
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(data_robust[:,0], distances)
for i in range(len(data_robust)):
    ax1.text(data_robust[i,0]+0.05, distances[i], labels[i], fontsize=12)
ax1.set_xlabel("Age (Robust)", fontsize=14)
ax1.set_ylabel("Distance to P1", fontsize=14)
ax1.set_xlim(min(data_robust[:,0])-0.5, max(data_robust[:,0])+0.5)
ax1.set_ylim(0, max(distances)+0.5)
ax1.tick_params(axis='both', which='major', labelsize=12)

# right: Distance vs Income (Robust normalized)
ax2 = plt.subplot(1, 2, 2)
ax2.scatter(data_robust[:,1], distances)
for i in range(len(data_robust)):
    ax2.text(data_robust[i,1]+0.05, distances[i], labels[i], fontsize=12)
ax2.set_xlabel("Income (Robust)", fontsize=14)
ax2.set_ylabel("Distance to P1", fontsize=14)
ax2.set_xlim(min(data_robust[:,1])-0.5, max(data_robust[:,1])+0.5)
ax2.set_ylim(0, max(distances)+0.5)
ax2.tick_params(axis='both', which='major', labelsize=12)

plt.tight_layout()
plt.show()
```
