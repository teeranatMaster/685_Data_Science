univariate Gaussian localtion and shape

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output

x = np.linspace(-10, 10, 400)
print(x)
for mu in np.linspace(-5, 5, 30):   
    clear_output(wait=True)         
    plt.figure(figsize=(6,4))       
    y = 1/(np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/2)
    plt.plot(x, y)
    plt.xlim(-10, 10)
    plt.ylim(0, 0.5)
    plt.title(f"Gaussian: mean={mu:.1f}, std=1")
    plt.show()
    time.sleep(0.2)                 
```

```python
from IPython.display import clear_output
import numpy as np
import matplotlib.pyplot as plt
import time

x = np.linspace(-10, 10, 400)

for sigma in np.linspace(0.5, 3, 30):   
    clear_output(wait=True)             
    plt.figure(figsize=(6,4))           
    y = 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-(x - 0)**2 / (2*sigma**2))
    plt.plot(x, y, color='tomato')
    plt.xlim(-10, 10)
    plt.ylim(0, 1.0)                    
    plt.title(f"Gaussian: mean=0, std={sigma:.2f}")
    plt.show()
    time.sleep(0.2)                     
```

Example of Guassian Distribtuion

```python
import numpy as np
import matplotlib.pyplot as plt

# random seed for reproducibility
np.random.seed(42)
n_trials = 100000

# uniform distribution: rolling a fair six-sided dice
single_rolls = np.random.randint(1, 7, size=n_trials)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(single_rolls, bins=np.arange(0.5,7.5,1), rwidth=0.8, density=True)
plt.xticks(range(1,7))
plt.xlabel('Dice face')
plt.ylabel('Probability')
plt.title('Single Dice Roll\n(Discrete Uniform)')

# sum of 30 dice rolls: approximate Gaussian
n_dice = 30
rolls_sum = np.sum(np.random.randint(1, 7, size=(n_trials, n_dice)), axis=1)

plt.subplot(1,2,2)
plt.hist(rolls_sum, bins=50, density=True, alpha=0.7, color='orange')
plt.xlabel('Sum of 30 dice')
plt.ylabel('Density')
plt.title('Sum of 30 Dice Rolls\n(Approx. Gaussian)')

plt.tight_layout()
plt.show()
```

Bivariate Gaussian Distribution Location and Shape

```python
from IPython.display import clear_output
import numpy as np
import matplotlib.pyplot as plt
import time

np.random.seed(0)
n = 1000          # number of points
cov = [[2, 0.5],  # fixed covariance matrix
       [0.5, 1]]

for mx in np.linspace(-5, 5, 30):
    clear_output(wait=True)
    plt.figure(figsize=(6,6))
    mu = [0, mx]   # mean vector changes along y-axis
    samples = np.random.multivariate_normal(mu, cov, n)
    plt.scatter(samples[:,0], samples[:,1], alpha=0.3, s=10)
    plt.xlim(-8, 8)
    plt.ylim(-5, 5)
    plt.title(f"2D Gaussian: mean=[{mx:.1f}, 0]")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
    time.sleep(0.2)
```

```python
from IPython.display import clear_output
import numpy as np
import matplotlib.pyplot as plt
import time

np.random.seed(0)
n = 1000
mu = [0, 0]         
rho = 0           # fixed correlation coefficient

# Let the variance of the second dimension change from 0.5 to 4
for var2 in np.linspace(0.5, 4.0, 30):
    clear_output(wait=True)
    plt.figure(figsize=(6,6))
    # construct covariance matrix:
    #  [[1, rho*sqrt(var2)],
    #   [rho*sqrt(var2), var2]] controls the variance of X2
    cov = [[1, rho*np.sqrt(var2)], 
           [rho*np.sqrt(var2), var2]]
    samples = np.random.multivariate_normal(mu, cov, n)
    plt.scatter(samples[:,0], samples[:,1], alpha=0.3, s=10, color='green')
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)
    plt.title(f"2D Gaussian: var(X2)={var2:.2f}, corr={rho}")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
    time.sleep(0.2)
```

```python
from IPython.display import clear_output
import numpy as np
import matplotlib.pyplot as plt
import time

np.random.seed(0)
n = 1000
mu = [0, 0]        # fixed mean

# covariance matrix parameter: from uncorrelated to strongly correlated
for rho in np.linspace(0, 0,95, 30):
    clear_output(wait=True)
    plt.figure(figsize=(6,6))
    # construct covariance matrix:
    #  [[1, rho],
    #   [rho, 1]] controls the correlation between the two dimensions
    cov = [[1, rho],
           [rho, 1]]
    samples = np.random.multivariate_normal(mu, cov, n)
    plt.scatter(samples[:,0], samples[:,1], alpha=0.3, s=10, color='tomato')
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.title(f"2D Gaussian: covariance off-diagonal={rho:.2f}")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
    time.sleep(0.2)
```
