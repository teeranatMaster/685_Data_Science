```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# Load the Wine dataset
wine = load_wine()

data = wine.data
targets = wine.target
feature_names = wine.feature_names
class_names = wine.target_names
```

Q1 Histogram

```python
# Task 1. Histogram on alcoho content

# Extract the 'alcohol' feature 
alcohol_content_index = feature_names.index('alcohol')
alcohol_content = data[:, alcohol_content_index]
# TODO: Implement your code here
fig,axs =  plt.subplots(1,3,figsize=(14,5),sharey=True)
for class_num in range(3):
    alcohol_content_filtered = alcohol_content[class_num==targets]
    axs[class_num].hist(alcohol_content_filtered,bins=10,edgecolor='black')
    axs[class_num].set_title(f"class: {class_num}")
    axs[class_num].set_xlabel(f"alcohol")
    axs[class_num].set_ylabel("frequency")
plt.show()
```

Q2 Box Plot

```python
# Task 2. Box plot on malic acid

# TODO: Implement your code here
fig,axs =  plt.subplots(1,3,figsize=(14,5))
malic_acid_index=  feature_names.index("malic_acid")
malic_acid =   data[:,malic_acid_index]
for class_num in range(3):
    malic_acid_filtered = malic_acid[class_num==targets]
    axs[class_num].boxplot(malic_acid_filtered,meanline=True)
    axs[class_num].set_title(f"class: {class_num}")
    axs[class_num].set_xlabel(f"malic_acid")
    axs[class_num].set_ylabel("frequency")
plt.show()
```

Q3 Scatter Plot 

```python
# Task 3 Scatter Plot between color intensity and magnesium

# TODO: Implement your code here
color_intensity_index=  feature_names.index("color_intensity")
magnesium_index= feature_names.index("magnesium")
color_intensity = data[:,color_intensity_index]
magnesium = data[:,magnesium_index]
fig,axs =  plt.subplots(1,1,figsize=(14,5))

# color intensity vs magnesium
colors = ['red','green','yellow']
labels = ['Class 0','Class 1','Class 2']

for class_num in range(3):
    color_intensity_filtered = color_intensity[class_num==targets]
    magnesium_filtered = magnesium[class_num==targets]
    axs.scatter(magnesium_filtered, color_intensity_filtered,color=colors[class_num], label=labels[class_num])
axs.set_title("Color Intensity vs. Magnesium")
axs.set_xlabel("Magnesium")
axs.set_ylabel("Color Intensity")
axs.legend()


```

```python
# Task 3 Scatter Plot between color intensity and malic acid

# TODO: Implement your code here
fig,axs =  plt.subplots(1,1,figsize=(14,5))
# color intensity vs Malic acid
for class_num in range(3):
    color_intensity_filtered = color_intensity[class_num==targets]
    malic_acid_filtered = malic_acid[class_num==targets]
    axs.scatter(malic_acid_filtered, color_intensity_filtered,color=colors[class_num], label=labels[class_num])
axs.set_title("Color Intensity vs. Malic Acid")
axs.set_xlabel("Malic Acid")
axs.set_ylabel("Color Intensity")
axs.legend()
```
