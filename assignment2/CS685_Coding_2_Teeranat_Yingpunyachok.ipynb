```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_wine
```

```python
wine_data = load_wine()
```

```python
X = wine_data.data
y = wine_data.target
y_class_nums = 3
```

```python
print(f"X shape:{X.shape}")
print(f"y shape:{y.shape}")
print(f" feature name:{wine_data.feature_names}")
print(type(wine_data.feature_names))
```

```python
print(X)
```

#### Find index of some features

```python
alcohol_index=  wine_data.feature_names.index("alcohol")
malic_acid_index=  wine_data.feature_names.index("malic_acid")
color_intensity_index=  wine_data.feature_names.index("color_intensity")
magnesium_index=  wine_data.feature_names.index("magnesium")
```

#### Get feature columns

```python
X_alcohol = X[:,alcohol_index]
X_malic = X[:,malic_acid_index]
X_color_intensity = X[:,color_intensity_index]
X_magnesium = X[:,magnesium_index]
```

## Q1 Histrogram

### #1,#2 histrogram graph

```python
fig,axs =  plt.subplots(1,3,figsize=(14,5),sharey=True)
for class_num in range(y_class_nums):
    X_alcohol_filtered = X_alcohol[y==class_num]
    axs[class_num].hist(X_alcohol_filtered,bins=10,edgecolor='black')
    axs[class_num].set_title(f"class: {class_num}")
    axs[class_num].set_xlabel(f"alcohol")
    axs[class_num].set_ylabel("frequency")
plt.show()
```

### #3 description

#### Class 0
- Range from 12.8 to 14.8
- Mean 13.74
- Median 13.75

#### Class 1
- Range from 11 to 14
- Mean 12.28
- Median 12.29

#### Class 2
- Range from 11 to 14
- Mean 13.15
- Median 13.16

#### Summary
- Class 0: have highest mean
- Class 1: have lowest mean, most frequency alcohol are around 12.25
- Class 2: Moderate mean, all the frequency spread out

## Q2 Box plot

### #1,#2 box plot graph

```python
fig,axs =  plt.subplots(1,3,figsize=(14,5))

for class_num in range(y_class_nums):
    X_malic_filtered = X_malic[y==class_num]
    axs[class_num].boxplot(X_malic_filtered,meanline=True)
    axs[class_num].set_title(f"class: {class_num}")
    axs[class_num].set_xlabel(f"malic_acid")
    axs[class_num].set_ylabel("frequency")
plt.show()
```

### #3 description

#### Highest Range
- class 1
#### Lowest Range
- class 0
#### Significent outliers
- class 0: many outlier compare to others two class
- class 1: there are some outliers but lower than class 0
- class 2: no outliers

## Q3 Scatter plot

### #1,#2 scatter graph

```python
fig,axs =  plt.subplots(1,2,figsize=(14,5))

# color intensity vs magnesium
colors = ['red','green','yellow']
labels = ['Class 0','Class 1','Class 2']

for class_num in range(y_class_nums):
    X_color_intensity_filtered = X_color_intensity[y==class_num]
    X_magnesium_filtered = X_magnesium[y==class_num]
    axs[0].scatter(X_magnesium_filtered, X_color_intensity_filtered,color=colors[class_num], label=labels[class_num])
axs[0].set_title("Color Intensity vs. Magnesium")
axs[0].set_xlabel("Magnesium")
axs[0].set_ylabel("Color Intensity")
axs[0].legend()

# color intensity vs Malic acid
for class_num in range(y_class_nums):
    X_color_intensity_filtered = X_color_intensity[y==class_num]
    X_malic_filtered = X_malic[y==class_num]
    axs[1].scatter(X_malic_filtered, X_color_intensity_filtered,color=colors[class_num], label=labels[class_num])
axs[1].set_title("Color Intensity vs. Malic Acid")
axs[1].set_xlabel("Malic Acid")
axs[1].set_ylabel("Color Intensity")
axs[1].legend()
```

### #3 description

#### Color Intensity vs. Magnesium
- class 0: all clustered in between class1 , and class2
- class 1: low color intensity in every magnesium level despite some is more then 130 
- class 2: higher color intensity at various magnesium levels
#### Color Intensity vs. Malic acid
- class 0: both malic acid and color intensity located at low zone
- class 1: mostly low malic acid and low color intensity
- class 2: spread across color intensity and malic acid

## Q4. Correlation Heatmap

### #1, #2 graph

```python
np.arange(len(wine_data.feature_names))
```

```python
correlation_matrix = np.corrcoef(X, rowvar=False)
print(correlation_matrix.shape)
print(f"correlation_matrix: {correlation_matrix}")
```

```python
fig,axs = plt.subplots(figsize=(10,8))

# axs.set_xtricks(np.arange()
axs.set_xticks(np.arange(len(wine_data.feature_names)))
axs.set_yticks(np.arange(len(wine_data.feature_names)))
axs.set_xticklabels(wine_data.feature_names,rotation=45,ha='right')
axs.set_yticklabels(wine_data.feature_names)

cs_m = axs.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)

for i in range(len(wine_data.feature_names)):
    for j in range(len(wine_data.feature_names)):
        label = np.round(correlation_matrix[i,j], decimals=2)
        axs.text(j, i, label, ha='center', va='center', color='black')

plt.title('Correlation Heatmap')

fig.colorbar(cs_m)
```

### #3 description

#### High positive correlation: flavanoids and total_phenols (+0.86)
#### High negative correlation: hue and malic_acid (-0.56)
